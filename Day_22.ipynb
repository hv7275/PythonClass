## 3. Step 1: Create Student Class

The **Student class** represents an individual student with:
- **Attributes:** name, roll_no, age, marks
- **Methods:** display details, update marks, update info

This class demonstrates **encapsulation** by protecting data with validation.


````python
# Step 1: Student Class
class Student:
    """Represents a single student with personal and academic information"""
    
    def __init__(self, name, roll_no, age, marks):
        """Initialize a student with basic information"""
        self.name = name
        self.roll_no = roll_no
        self.age = age
        self.marks = marks
    
    def display_details(self):
        """Display student information in formatted way"""
        print(f"{'â”€' * 50}")
        print(f"Name       : {self.name}")
        print(f"Roll No    : {self.roll_no}")
        print(f"Age        : {self.age}")
        print(f"Marks      : {self.marks}")
        print(f"{'â”€' * 50}")
    
    def update_marks(self, new_marks):
        """Update student marks with validation"""
        if 0 <= new_marks <= 100:
            self.marks = new_marks
            print(f"âœ“ Marks updated to {new_marks} for {self.name}")
            return True
        else:
            print(f"âŒ Invalid marks! Must be between 0 and 100")
            return False
    
    def update_info(self, name=None, age=None, roll_no=None):
        """Update student information"""
        if name:
            self.name = name
        if age:
            if 5 < age < 100:
                self.age = age
            else:
                print("âŒ Invalid age!")
                return False
        if roll_no:
            self.roll_no = roll_no
        print(f"âœ“ Information updated for {self.name}")
        return True
    
    def __str__(self):
        """String representation of student"""
        return f"{self.roll_no:4} | {self.name:15} | Age: {self.age:2} | Marks: {self.marks:3}"

# Test the Student Class
print("="*50)
print("Testing Student Class")
print("="*50)

student1 = Student("Alice", 101, 18, 85)
student1.display_details()

student1.update_marks(90)
student1.update_info(age=19)

print("\nUpdated Details:")
student1.display_details()
````

## 4. Step 2: Create StudentManager Class

The **StudentManager class** manages multiple students and implements CRUD operations:
- **Create:** add_student()
- **Read:** search_student(), display_all()
- **Update:** update_student()
- **Delete:** remove_student()

This class demonstrates **collection management** and **responsibility separation**.


````python
# Step 2: StudentManager Class
class StudentManager:
    """Manages multiple students and performs CRUD operations"""
    
    def __init__(self):
        """Initialize with empty student list"""
        self.students = []
    
    def add_student(self, student):
        """Add a new student to the system"""
        # Check if roll number already exists
        for s in self.students:
            if s.roll_no == student.roll_no:
                print(f"âŒ Student with Roll No {student.roll_no} already exists!")
                return False
        
        self.students.append(student)
        print(f"âœ“ Student {student.name} added successfully!")
        return True
    
    def remove_student(self, roll_no):
        """Remove student by roll number"""
        for i, student in enumerate(self.students):
            if student.roll_no == roll_no:
                name = student.name
                self.students.pop(i)
                print(f"âœ“ Student {name} removed successfully!")
                return True
        
        print(f"âŒ Student with Roll No {roll_no} not found!")
        return False
    
    def search_student(self, roll_no):
        """Search student by roll number"""
        for student in self.students:
            if student.roll_no == roll_no:
                return student
        return None
    
    def update_student(self, roll_no, name=None, age=None, marks=None):
        """Update student information"""
        student = self.search_student(roll_no)
        
        if student is None:
            print(f"âŒ Student with Roll No {roll_no} not found!")
            return False
        
        if name:
            student.name = name
        if age:
            if 5 < age < 100:
                student.age = age
            else:
                print("âŒ Invalid age!")
                return False
        if marks:
            if not student.update_marks(marks):
                return False
        
        print(f"âœ“ Student {student.name} updated successfully!")
        return True
    
    def display_all(self):
        """Display all students in tabular format"""
        if not self.students:
            print("âŒ No students in the system!")
            return
        
        print("\n" + "="*60)
        print(f"{'Roll No':^8} | {'Name':^20} | {'Age':^5} | {'Marks':^6}")
        print("="*60)
        
        for student in self.students:
            print(student)
        
        print("="*60 + "\n")
    
    def display_student(self, roll_no):
        """Display details of a specific student"""
        student = self.search_student(roll_no)
        
        if student is None:
            print(f"âŒ Student with Roll No {roll_no} not found!")
            return False
        
        student.display_details()
        return True
    
    def get_total_students(self):
        """Get total number of students"""
        return len(self.students)

# Test StudentManager
print("\n" + "="*50)
print("Testing StudentManager Class")
print("="*50 + "\n")

manager = StudentManager()

# Add students
s1 = Student("Alice", 101, 18, 85)
s2 = Student("Bob", 102, 19, 78)
s3 = Student("Charlie", 103, 18, 92)

manager.add_student(s1)
manager.add_student(s2)
manager.add_student(s3)

# Display all
manager.display_all()

# Search
print("Searching for Roll No 102:")
manager.display_student(102)

# Update
manager.update_student(101, marks=95)

# Display all again
manager.display_all()
````

## 5. Step 3: Implement Menu-Based System

The **menu system** provides an interactive interface for users to interact with the system.
It uses a `while True` loop and displays options for each operation.


````python
# Step 3: Menu-Based System
def display_menu():
    """Display main menu options"""
    print("\n" + "="*50)
    print("ğŸ“š STUDENT MANAGEMENT SYSTEM")
    print("="*50)
    print("1. â• Add Student")
    print("2. âŒ Remove Student")
    print("3. âœï¸  Update Student")
    print("4. ğŸ” Search Student")
    print("5. ğŸ“‹ Display All Students")
    print("6. ğŸšª Exit")
    print("="*50)

def menu_system():
    """Interactive menu for student management"""
    manager = StudentManager()
    
    while True:
        display_menu()
        choice = input("Enter your choice (1-6): ").strip()
        
        if choice == "1":
            # Add Student
            print("\n--- Add New Student ---")
            try:
                name = input("Enter student name: ").strip()
                roll_no = int(input("Enter roll number: "))
                age = int(input("Enter age: "))
                marks = int(input("Enter marks (0-100): "))
                
                if marks < 0 or marks > 100:
                    print("âŒ Marks must be between 0 and 100!")
                    continue
                
                student = Student(name, roll_no, age, marks)
                manager.add_student(student)
            
            except ValueError:
                print("âŒ Invalid input! Please enter correct data types.")
        
        elif choice == "2":
            # Remove Student
            print("\n--- Remove Student ---")
            try:
                roll_no = int(input("Enter roll number to remove: "))
                manager.remove_student(roll_no)
            except ValueError:
                print("âŒ Invalid roll number!")
        
        elif choice == "3":
            # Update Student
            print("\n--- Update Student ---")
            try:
                roll_no = int(input("Enter roll number to update: "))
                print("(Press Enter to skip any field)")
                
                name = input("New name (or press Enter to skip): ").strip() or None
                age_input = input("New age (or press Enter to skip): ").strip()
                age = int(age_input) if age_input else None
                marks_input = input("New marks (or press Enter to skip): ").strip()
                marks = int(marks_input) if marks_input else None
                
                manager.update_student(roll_no, name=name, age=age, marks=marks)
            
            except ValueError:
                print("âŒ Invalid input! Please enter correct data types.")
        
        elif choice == "4":
            # Search Student
            print("\n--- Search Student ---")
            try:
                roll_no = int(input("Enter roll number to search: "))
                manager.display_student(roll_no)
            except ValueError:
                print("âŒ Invalid roll number!")
        
        elif choice == "5":
            # Display All
            print("\n--- All Students ---")
            manager.display_all()
            print(f"Total Students: {manager.get_total_students()}")
        
        elif choice == "6":
            # Exit
            print("\nğŸ‘‹ Thank you for using Student Management System!")
            print("Goodbye!\n")
            break
        
        else:
            print("âŒ Invalid choice! Please enter 1-6.")

# Demo of menu system (commented out as it requires user input)
# Uncomment and run in terminal for interactive use
# menu_system()

print("âœ“ Menu system defined successfully!")
print("To use the interactive menu, run: menu_system()")
````

## 7. Bonus Features (Optional)

Enhance the system with additional functionality:


````python
# Bonus Features: Enhanced StudentManager
class EnhancedStudentManager(StudentManager):
    """Extended StudentManager with sorting and file operations"""
    
    def sort_by_marks(self, descending=True):
        """Sort students by marks"""
        sorted_students = sorted(self.students, 
                                key=lambda s: s.marks, 
                                reverse=descending)
        
        print("\n" + "="*60)
        print(f"Students sorted by Marks ({'Highest to Lowest' if descending else 'Lowest to Highest'})")
        print("="*60)
        print(f"{'Rank':^6} | {'Roll No':^8} | {'Name':^20} | {'Marks':^6}")
        print("="*60)
        
        for rank, student in enumerate(sorted_students, 1):
            print(f"{rank:^6} | {student.roll_no:^8} | {student.name:^20} | {student.marks:^6}")
        print("="*60)
    
    def sort_by_name(self):
        """Sort students by name alphabetically"""
        sorted_students = sorted(self.students, 
                                key=lambda s: s.name)
        
        print("\n" + "="*60)
        print("Students sorted by Name (A-Z)")
        print("="*60)
        print(f"{'Roll No':^8} | {'Name':^20} | {'Age':^5} | {'Marks':^6}")
        print("="*60)
        
        for student in sorted_students:
            print(student)
        print("="*60)
    
    def get_topper(self):
        """Get student with highest marks"""
        if not self.students:
            print("âŒ No students in the system!")
            return None
        
        topper = max(self.students, key=lambda s: s.marks)
        print(f"\nğŸ† TOPPER: {topper.name} (Roll No: {topper.roll_no}) - Marks: {topper.marks}")
        return topper
    
    def save_to_file(self, filename="students.txt"):
        """Save student data to file"""
        try:
            with open(filename, 'w') as f:
                f.write("="*60 + "\n")
                f.write("STUDENT MANAGEMENT SYSTEM - DATA BACKUP\n")
                f.write("="*60 + "\n\n")
                
                f.write(f"{'Roll No':^8} | {'Name':^20} | {'Age':^5} | {'Marks':^6}\n")
                f.write("-"*60 + "\n")
                
                for student in self.students:
                    f.write(f"{student.roll_no:^8} | {student.name:^20} | {student.age:^5} | {student.marks:^6}\n")
            
            print(f"âœ“ Data saved to {filename}")
            return True
        except Exception as e:
            print(f"âŒ Error saving file: {e}")
            return False
    
    def load_from_file(self, filename="students.txt"):
        """Load student data from file"""
        try:
            with open(filename, 'r') as f:
                lines = f.readlines()
                # Skip header lines
                data_lines = [l for l in lines if l.strip() and 'â”‚' in l and 'Roll No' not in l and '=' not in l and '-' not in l]
                
                for line in data_lines:
                    parts = [p.strip() for p in line.split('â”‚')]
                    if len(parts) >= 4:
                        try:
                            roll_no = int(parts[0])
                            name = parts[1]
                            age = int(parts[2])
                            marks = int(parts[3])
                            student = Student(name, roll_no, age, marks)
                            self.add_student(student)
                        except ValueError:
                            continue
            
            print(f"âœ“ Data loaded from {filename}")
            return True
        except FileNotFoundError:
            print(f"âŒ File {filename} not found!")
            return False

# Test Bonus Features
print("\n" + "="*60)
print("Testing Bonus Features")
print("="*60)

enhanced_manager = EnhancedStudentManager()

# Add test students
students = [
    Student("Alice Johnson", 101, 18, 95),
    Student("Bob Smith", 102, 19, 78),
    Student("Charlie Brown", 103, 18, 92),
    Student("Diana Prince", 104, 20, 88),
    Student("Eve Wilson", 105, 19, 85)
]

for student in students:
    enhanced_manager.add_student(student)

# Sort by marks
enhanced_manager.sort_by_marks(descending=True)

# Sort by name
enhanced_manager.sort_by_name()

# Get topper
enhanced_manager.get_topper()

print("\nâœ“ Bonus features demonstrated!")
````

````python
# Mini Project Extension: Teacher Management System
class Teacher:
    """Represents a teacher who manages students"""
    
    def __init__(self, teacher_id, name, subject):
        """Initialize teacher information"""
        self.teacher_id = teacher_id
        self.name = name
        self.subject = subject
        self.assigned_students = []
    
    def assign_student(self, student):
        """Assign a student to this teacher"""
        if student not in self.assigned_students:
            self.assigned_students.append(student)
            print(f"âœ“ {student.name} assigned to {self.name} for {self.subject}")
            return True
        else:
            print(f"âŒ {student.name} already assigned to {self.name}")
            return False
    
    def set_marks(self, student, marks):
        """Set marks for a student"""
        if student in self.assigned_students:
            student.update_marks(marks)
            print(f"âœ“ {self.name} set marks {marks} for {student.name} in {self.subject}")
            return True
        else:
            print(f"âŒ {student.name} not assigned to {self.name}")
            return False
    
    def display_class(self):
        """Display all assigned students"""
        print(f"\n{'='*60}")
        print(f"Class: {self.subject} | Teacher: {self.name}")
        print(f"{'='*60}")
        print(f"{'Roll No':^8} | {'Name':^20} | {'Marks':^8}")
        print("-"*60)
        
        if not self.assigned_students:
            print("No students assigned yet.")
        else:
            for student in self.assigned_students:
                print(f"{student.roll_no:^8} | {student.name:^20} | {student.marks:^8}")
        print("="*60)

# Test Teacher Management System
print("\n" + "="*70)
print("MINI PROJECT: Teacher Management System")
print("="*70)

# Create teachers
math_teacher = Teacher(1, "Mr. Sharma", "Mathematics")
science_teacher = Teacher(2, "Mrs. Patel", "Science")

# Create students
students_list = [
    Student("Aisha Khan", 301, 17, 0),
    Student("Rohan Gupta", 302, 18, 0),
    Student("Sana Ali", 303, 17, 0)
]

# Assign students to teachers
print("\n--- Assigning Students to Teachers ---")
for student in students_list:
    math_teacher.assign_student(student)

for student in students_list[:2]:  # Assign first 2 to science teacher
    science_teacher.assign_student(student)

# Teachers set marks
print("\n--- Teachers Setting Marks ---")
math_teacher.set_marks(students_list[0], 88)
math_teacher.set_marks(students_list[1], 92)
math_teacher.set_marks(students_list[2], 85)

science_teacher.set_marks(students_list[0], 91)
science_teacher.set_marks(students_list[1], 87)

# Display classes
math_teacher.display_class()
science_teacher.display_class()

print("\nâœ“ Teacher Management System successfully integrated!")
````
