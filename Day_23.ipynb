## 2. What is Modular Programming?

**Modular Programming** is an approach where large programs are divided into separate, independent modules.

### Key Principles:

1. **Separation of Concerns** - Each module has one responsibility
2. **Independence** - Modules work independently
3. **Encapsulation** - Internal details hidden, clean interface
4. **Reusability** - Modules can be used in other projects

## 3. Creating Modules in Python

In Python, **any `.py` file is a module**!

### Creating a Module:

```python
# student.py - This IS a module
class Student:
    pass

def validate_age(age):
    return 0 < age < 150
```

### Importing Modules:

```python
# main.py - Import from student module
import student
from student import Student, validate_age
from student import Student as S
```

### Different Import Styles:

```python
# Style 1: Import entire module
import student
student.Student("Alice", 101, 18, 85)

# Style 2: Import specific class/function
from student import Student
Student("Alice", 101, 18, 85)

# Style 3: Import multiple items
from student import Student, validate_age

# Style 4: Import with alias
from student import Student as StudentClass
from manager import StudentManager as SMS

# Style 5: Import all (use carefully!)
from student import *
```

### Module Search Path:

Python looks for modules in this order:
1. Current directory (same folder)
2. Environment variable PYTHONPATH
3. Installation-dependent default paths
4. Built-in modules

### Key Points:
âœ… Module name = filename (without .py)  
âœ… Any Python file can be a module  
âœ… Import runs the module code  
âœ… Use clear, descriptive names  


## 4. Step 1: Create student.py Module

This module contains the **Student class** - the data model for our system.


````python
# Content of student.py module
student_module_code = '''
"""
student.py - Student data model module
Contains the Student class for representing individual students
"""

class Student:
    """Represents a single student with personal and academic information"""
    
    def __init__(self, name, roll_no, age, marks):
        """Initialize a student with basic information"""
        self.name = name
        self.roll_no = roll_no
        self.age = age
        self.marks = marks
    
    def display(self):
        """Display student information in formatted way"""
        print(f"{'â”€' * 50}")
        print(f"Name       : {self.name}")
        print(f"Roll No    : {self.roll_no}")
        print(f"Age        : {self.age}")
        print(f"Marks      : {self.marks}")
        print(f"{'â”€' * 50}")
    
    def update_marks(self, new_marks):
        """Update student marks with validation"""
        if 0 <= new_marks <= 100:
            self.marks = new_marks
            print(f"âœ“ Marks updated to {new_marks} for {self.name}")
            return True
        else:
            print(f"âŒ Invalid marks! Must be between 0 and 100")
            return False
    
    def update_info(self, name=None, age=None):
        """Update student information"""
        if name:
            self.name = name
        if age:
            if 5 < age < 100:
                self.age = age
            else:
                print("âŒ Invalid age!")
                return False
        print(f"âœ“ Information updated for {self.name}")
        return True
    
    def __str__(self):
        """String representation of student"""
        return f"{self.roll_no:4} | {self.name:15} | Age: {self.age:2} | Marks: {self.marks:3}"


# Helper functions for validation
def validate_age(age):
    """Validate age input"""
    return 5 < age < 100

def validate_marks(marks):
    """Validate marks input"""
    return 0 <= marks <= 100

def validate_roll_no(roll_no):
    """Validate roll number (must be positive integer)"""
    return isinstance(roll_no, int) and roll_no > 0
'''

# Show the code
print("="*70)
print("STUDENT.PY MODULE CODE")
print("="*70)
print(student_module_code)
print("="*70)
print("\nThis module would be saved as: student.py")
````

## 5. Step 2: Create manager.py Module

This module contains the **StudentManager class** - handles all business logic and operations.


````python
# Content of manager.py module
manager_module_code = '''
"""
manager.py - Student management module
Contains the StudentManager class for handling CRUD operations
"""

# In real project, this would be: from student import Student
# For this demo, we'll define it inline

class StudentManager:
    """Manages multiple students and performs CRUD operations"""
    
    def __init__(self):
        """Initialize with empty student list"""
        self.students = []
    
    def add_student(self, student):
        """Add a new student to the system"""
        # Check if roll number already exists
        for s in self.students:
            if s.roll_no == student.roll_no:
                print(f"âŒ Student with Roll No {student.roll_no} already exists!")
                return False
        
        self.students.append(student)
        print(f"âœ“ Student {student.name} added successfully!")
        return True
    
    def remove_student(self, roll_no):
        """Remove student by roll number"""
        for i, student in enumerate(self.students):
            if student.roll_no == roll_no:
                name = student.name
                self.students.pop(i)
                print(f"âœ“ Student {name} removed successfully!")
                return True
        
        print(f"âŒ Student with Roll No {roll_no} not found!")
        return False
    
    def search_student(self, roll_no):
        """Search student by roll number"""
        for student in self.students:
            if student.roll_no == roll_no:
                return student
        return None
    
    def update_student(self, roll_no, name=None, age=None, marks=None):
        """Update student information"""
        student = self.search_student(roll_no)
        
        if student is None:
            print(f"âŒ Student with Roll No {roll_no} not found!")
            return False
        
        if name:
            student.name = name
        if age:
            if 5 < age < 100:
                student.age = age
            else:
                print("âŒ Invalid age!")
                return False
        if marks:
            if not student.update_marks(marks):
                return False
        
        print(f"âœ“ Student {student.name} updated successfully!")
        return True
    
    def display_all(self):
        """Display all students in tabular format"""
        if not self.students:
            print("âŒ No students in the system!")
            return
        
        print("\\n" + "="*60)
        print(f"{'Roll No':^8} | {'Name':^20} | {'Age':^5} | {'Marks':^6}")
        print("="*60)
        
        for student in self.students:
            print(student)
        
        print("="*60 + "\\n")
    
    def display_student(self, roll_no):
        """Display details of a specific student"""
        student = self.search_student(roll_no)
        
        if student is None:
            print(f"âŒ Student with Roll No {roll_no} not found!")
            return False
        
        student.display()
        return True
    
    def get_total_students(self):
        """Get total number of students"""
        return len(self.students)
'''

print("="*70)
print("MANAGER.PY MODULE CODE")
print("="*70)
print(manager_module_code)
print("="*70)
print("\nThis module would be saved as: manager.py")
print("\nNote: In a real project, manager.py would start with:")
print("  from student import Student")
````

## 6. Step 3: Create main.py (Controller)

This is the **entry point** of the application - handles user interface and interactions.


````python
# Content of main.py module
main_module_code = '''
"""
main.py - Main application controller
Entry point of the Student Management System
Imports and uses student and manager modules
"""

# Import from our modules
from student import Student
from manager import StudentManager

def display_menu():
    """Display main menu options"""
    print("\\n" + "="*50)
    print("ğŸ“š STUDENT MANAGEMENT SYSTEM")
    print("="*50)
    print("1. â• Add Student")
    print("2. âŒ Remove Student")
    print("3. âœï¸  Update Student")
    print("4. ğŸ” Search Student")
    print("5. ğŸ“‹ Display All Students")
    print("6. ğŸšª Exit")
    print("="*50)

def main():
    """Main program flow"""
    manager = StudentManager()
    
    # Add some sample students
    sample_students = [
        Student("Alice Johnson", 101, 18, 85),
        Student("Bob Smith", 102, 19, 78),
        Student("Charlie Brown", 103, 18, 92)
    ]
    
    for student in sample_students:
        manager.add_student(student)
    
    while True:
        display_menu()
        choice = input("Enter your choice (1-6): ").strip()
        
        if choice == "1":
            print("\\n--- Add New Student ---")
            try:
                name = input("Enter student name: ").strip()
                roll_no = int(input("Enter roll number: "))
                age = int(input("Enter age: "))
                marks = int(input("Enter marks (0-100): "))
                
                if marks < 0 or marks > 100:
                    print("âŒ Marks must be between 0 and 100!")
                    continue
                
                student = Student(name, roll_no, age, marks)
                manager.add_student(student)
            
            except ValueError:
                print("âŒ Invalid input!")
        
        elif choice == "2":
            print("\\n--- Remove Student ---")
            try:
                roll_no = int(input("Enter roll number to remove: "))
                manager.remove_student(roll_no)
            except ValueError:
                print("âŒ Invalid roll number!")
        
        elif choice == "3":
            print("\\n--- Update Student ---")
            try:
                roll_no = int(input("Enter roll number to update: "))
                name = input("New name (or press Enter to skip): ").strip() or None
                age_input = input("New age (or press Enter to skip): ").strip()
                age = int(age_input) if age_input else None
                marks_input = input("New marks (or press Enter to skip): ").strip()
                marks = int(marks_input) if marks_input else None
                
                manager.update_student(roll_no, name=name, age=age, marks=marks)
            
            except ValueError:
                print("âŒ Invalid input!")
        
        elif choice == "4":
            print("\\n--- Search Student ---")
            try:
                roll_no = int(input("Enter roll number to search: "))
                manager.display_student(roll_no)
            except ValueError:
                print("âŒ Invalid roll number!")
        
        elif choice == "5":
            print("\\n--- All Students ---")
            manager.display_all()
            print(f"Total Students: {manager.get_total_students()}")
        
        elif choice == "6":
            print("\\nğŸ‘‹ Thank you for using Student Management System!")
            break
        
        else:
            print("âŒ Invalid choice! Please enter 1-6.")

if __name__ == "__main__":
    main()
'''

print("="*70)
print("MAIN.PY MODULE CODE")
print("="*70)
print(main_module_code)
print("="*70)
print("\nThis module would be saved as: main.py")
print("\nKey points:")
print("  âœ“ Imports from other modules (student, manager)")
print("  âœ“ Entry point with if __name__ == '__main__'")
print("  âœ“ Uses classes from imported modules")
````

## 9. Handling Errors in Multi-Module Projects

Robust error handling is crucial when working with multiple modules.


````python
# Error Handling in Multi-Module Projects
print("\n" + "="*70)
print("ERROR HANDLING IN MULTI-MODULE PROJECTS")
print("="*70)

error_handling_code = '''
# 1. Try-except when importing
â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
try:
    from student import Student
    print("âœ“ Student module loaded")
except ImportError as e:
    print(f"âŒ Cannot import Student: {e}")
    # Handle gracefully - use default or exit
    exit(1)


# 2. Try-except for file operations
â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def load_student_data(filename):
    try:
        with open(filename, 'r') as f:
            data = f.read()
        return data
    except FileNotFoundError:
        print(f"âŒ File {filename} not found!")
        return None
    except IOError as e:
        print(f"âŒ Error reading file: {e}")
        return None


# 3. Validation for input data
â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def validate_student_data(name, roll_no, age, marks):
    """Validate student data before creating Student object"""
    
    # Check name
    if not name or not isinstance(name, str):
        raise ValueError("âŒ Name must be a non-empty string")
    
    # Check roll number
    if not isinstance(roll_no, int) or roll_no <= 0:
        raise ValueError("âŒ Roll number must be positive integer")
    
    # Check age
    if not isinstance(age, int) or age < 5 or age > 100:
        raise ValueError("âŒ Age must be between 5 and 100")
    
    # Check marks
    if not isinstance(marks, int) or marks < 0 or marks > 100:
        raise ValueError("âŒ Marks must be between 0 and 100")
    
    return True


# 4. Try-except in manager operations
â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def safe_add_student(manager, name, roll_no, age, marks):
    """Safely add student with error handling"""
    try:
        # Validate data first
        validate_student_data(name, roll_no, age, marks)
        
        # Create student
        from student import Student
        student = Student(name, roll_no, age, marks)
        
        # Add to manager
        if manager.add_student(student):
            return True
        else:
            print("âŒ Failed to add student (might be duplicate)")
            return False
    
    except ValueError as e:
        print(f"âŒ Validation error: {e}")
        return False
    except ImportError as e:
        print(f"âŒ Module import error: {e}")
        return False
    except Exception as e:
        print(f"âŒ Unexpected error: {e}")
        return False


# 5. Custom exceptions for modules
â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
class StudentError(Exception):
    """Custom exception for student-related errors"""
    pass

class ManagerError(Exception):
    """Custom exception for manager-related errors"""
    pass

def add_with_custom_exception(manager, student):
    try:
        if not manager.add_student(student):
            raise ManagerError("Student with this roll number already exists")
        return True
    except ManagerError as e:
        print(f"Manager Error: {e}")
        return False


# 6. Logging errors (recommended for production)
â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
import logging

logging.basicConfig(filename='errors.log', level=logging.ERROR)

try:
    # Your code here
    pass
except Exception as e:
    logging.error(f"Error occurred: {e}", exc_info=True)
    print("âŒ Error logged. Check errors.log for details")
'''

print(error_handling_code)

print("\n" + "="*70)
print("ERROR HANDLING CHECKLIST")
print("="*70)
checklist = """
âœ“ Import errors - Use try-except for module imports
âœ“ File errors - Handle FileNotFoundError, IOError
âœ“ Data validation - Check types and ranges before use
âœ“ Type errors - Validate input types before processing
âœ“ Logical errors - Handle edge cases and invalid operations
âœ“ User input - Never trust user input, always validate
âœ“ Database/File operations - Use try-except blocks
âœ“ Logging - Log errors for debugging
âœ“ Graceful degradation - Don't crash, handle failures
âœ“ Error messages - Provide clear, actionable error messages
"""
print(checklist)
````

````python
# Storage Module Example
storage_module_code = '''
"""
storage.py - File storage module
Handles saving and loading student data using JSON format
"""

import json
import os

class StorageManager:
    """Manages file storage and retrieval of student data"""
    
    def __init__(self, data_dir="data"):
        """Initialize storage manager"""
        self.data_dir = data_dir
        self.students_file = os.path.join(data_dir, "students.json")
        
        # Create data directory if it doesn't exist
        if not os.path.exists(data_dir):
            os.makedirs(data_dir)
    
    def save_students(self, manager):
        """Save all students to JSON file"""
        try:
            students_data = []
            
            for student in manager.students:
                student_dict = {
                    "name": student.name,
                    "roll_no": student.roll_no,
                    "age": student.age,
                    "marks": student.marks
                }
                students_data.append(student_dict)
            
            # Write to JSON file
            with open(self.students_file, 'w') as f:
                json.dump(students_data, f, indent=4)
            
            print(f"âœ“ Data saved to {self.students_file}")
            print(f"  Total students saved: {len(students_data)}")
            return True
        
        except IOError as e:
            print(f"âŒ Error saving file: {e}")
            return False
    
    def load_students(self, manager):
        """Load students from JSON file"""
        try:
            if not os.path.exists(self.students_file):
                print(f"âš ï¸  No saved data found at {self.students_file}")
                return False
            
            with open(self.students_file, 'r') as f:
                students_data = json.load(f)
            
            # Load into manager
            from student import Student
            
            for student_dict in students_data:
                student = Student(
                    name=student_dict["name"],
                    roll_no=student_dict["roll_no"],
                    age=student_dict["age"],
                    marks=student_dict["marks"]
                )
                manager.add_student(student)
            
            print(f"âœ“ Data loaded from {self.students_file}")
            print(f"  Total students loaded: {len(students_data)}")
            return True
        
        except IOError as e:
            print(f"âŒ Error loading file: {e}")
            return False
        except json.JSONDecodeError as e:
            print(f"âŒ Error reading JSON: {e}")
            return False

# Usage in main.py:
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# from storage import StorageManager
# from manager import StudentManager
# 
# manager = StudentManager()
# storage = StorageManager()
# 
# # Load existing data
# storage.load_students(manager)
# 
# # ... user interactions ...
# 
# # Save data before exit
# storage.save_students(manager)
'''

print("="*70)
print("STORAGE.PY MODULE CODE")
print("="*70)
print(storage_module_code)

print("\n" + "="*70)
print("JSON FILE EXAMPLE (data/students.json)")
print("="*70)

json_example = '''
[
    {
        "name": "Alice Johnson",
        "roll_no": 101,
        "age": 18,
        "marks": 85
    },
    {
        "name": "Bob Smith",
        "roll_no": 102,
        "age": 19,
        "marks": 78
    },
    {
        "name": "Charlie Brown",
        "roll_no": 103,
        "age": 18,
        "marks": 92
    }
]
'''

print(json_example)

print("\n" + "="*70)
print("INTEGRATION WITH MAIN.PY")
print("="*70)

integration_code = '''
# Modified main.py with storage

from student import Student
from manager import StudentManager
from storage import StorageManager

def main():
    manager = StudentManager()
    storage = StorageManager("data")  # Creates data/ folder
    
    # Load existing data on startup
    storage.load_students(manager)
    
    while True:
        # ... menu and operations ...
        
        if choice == "6":  # Before exit
            save_choice = input("Save data before exit? (y/n): ").lower()
            if save_choice == 'y':
                storage.save_students(manager)
            print("Goodbye!")
            break

if __name__ == "__main__":
    main()
'''

print(integration_code)
````
