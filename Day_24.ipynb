## 1. Introduction

### What is a Python Package?

A **package** is a way of organizing related modules into a directory hierarchy. It's a directory that contains Python modules and a special `__init__.py` file.

**Simple Definition:**
- **Module:** A single `.py` file
- **Package:** A folder containing multiple `.py` files and an `__init__.py` file

### Module vs Package:

| Aspect | Module | Package |
|--------|--------|---------|
| **Type** | Single `.py` file | Directory with `.py` files |
| **Example** | `math_utils.py` | `my_package/` folder |
| **Contents** | Functions, classes | Multiple modules |
| **Import** | `import math_utils` | `import my_package` |
| **Structure** | One file | Nested hierarchy |

### Why Use Packages?

âœ… **Organization** - Group related modules together  
âœ… **Reusability** - Use in multiple projects  
âœ… **Scalability** - Easy to extend functionality  
âœ… **Distribution** - Share via pip install  
âœ… **Namespacing** - Avoid naming conflicts  
âœ… **Professional** - Industry standard structure  

### Real-World Examples:

```python
import numpy              # numpy package
import pandas            # pandas package
import requests          # requests package
from flask import Flask  # Flask package with Flask module
```

These are all packages with multiple modules inside!


## 2. Package Structure

A Python package is simply **a directory with an `__init__.py` file**.

### Basic Package Structure:

```
mypackage/                    # Package folder
â”œâ”€â”€ __init__.py              # Makes it a package (can be empty)
â”œâ”€â”€ module1.py               # Module 1
â”œâ”€â”€ module2.py               # Module 2
â””â”€â”€ subpackage/              # Sub-package (nested)
    â”œâ”€â”€ __init__.py
    â””â”€â”€ module3.py
```

### Modern Python (3.3+):

Python 3.3 introduced **namespace packages** that don't need `__init__.py`:

```
mypackage/                    # Still a package!
â”œâ”€â”€ module1.py               # No __init__.py needed
â”œâ”€â”€ module2.py
â””â”€â”€ subpackage/
    â””â”€â”€ module3.py
```

However, **it's still best practice to include `__init__.py`** for clarity and compatibility.

### File Responsibilities:

| File | Purpose |
|------|---------|
| `__init__.py` | Marks directory as package, can import contents |
| `module1.py` | Contains functions/classes |
| `module2.py` | Contains functions/classes |
| `subpackage/` | Nested package with its own `__init__.py` |

### Visual Hierarchy:

```
my_app/
â”‚
â”œâ”€â”€ main.py                 # Main application
â”‚
â””â”€â”€ my_package/             # Our package
```

When you do `import my_package`, Python looks for:
1. `my_package/__init__.py` first
2. Then loads modules as needed

## 5. Using `__init__.py` - Making Imports Easier

The `__init__.py` file allows you to simplify imports and expose the package's public API.

### `__init__.py` Purposes:

1. **Mark directory as package** - Makes Python recognize it as a package
2. **Import commonly used items** - Expose functions at package level
3. **Initialize package** - Run setup code
4. **Control `__all__`** - Define what `import *` includes

### Examples of `__init__.py`:

**Option 1: Empty (simplest)**
```python
# File: my_package/__init__.py
# Just marks the folder as a package
```

**Option 2: Import common items**
```python
# File: my_package/__init__.py
from .calculator import add, subtract, multiply
from .formatter import greet, uppercase
from .utils import is_even, is_odd
```

**Option 3: With package metadata**
```python
# File: my_package/__init__.py
__version__ = "1.0.0"
__author__ = "Your Name"

from .calculator import add, subtract
from .formatter import greet
```

**Option 4: With `__all__`**
```python
# File: my_package/__init__.py
__all__ = ['add', 'greet', 'is_even']

from .calculator import add
from .formatter import greet
from .utils import is_even
```

### How It Simplifies Imports:

**Without smart `__init__.py`:**
```python
from my_package.calculator import add
from my_package.formatter import greet
from my_package.utils import is_even
```

**With smart `__init__.py`:**
```python
from my_package import add, greet, is_even
```

Much cleaner! âœ¨

## 7. Packaging with setup.py - Distribution Basics

The `setup.py` file makes your package installable via `pip`.

### What is setup.py?

`setup.py` is a configuration file that:
- Describes your package
- Specifies dependencies
- Defines how to install it
- Enables distribution to PyPI

### Basic setup.py Structure:

```python
from setuptools import setup, find_packages

setup(
    name="my_package",
    version="1.0.0",
    description="A useful package for utilities",
    author="Your Name",
    packages=find_packages(),
)
```

### Full setup.py with all details:

```python
from setuptools import setup, find_packages

setup(
    name="my_package",
    version="1.0.0",
    description="A short description",
    long_description="A longer description",
    author="Your Name",
    author_email="your@email.com",
    url="https://github.com/yourname/my_package",
    license="MIT",
    packages=find_packages(),
    install_requires=[
        "numpy>=1.0",
        "requests>=2.0",
    ],
    classifiers=[
        "Programming Language :: Python :: 3",
        "License :: OSI Approved :: MIT License",
        "Operating System :: OS Independent",
    ],
    python_requires=">=3.6",
)
```

### Key Parameters:

| Parameter | Meaning |
|-----------|---------|
| `name` | Package name (used in pip install) |
| `version` | Package version |
| `description` | Short description |
| `author` | Your name |
| `packages` | Folders to include (use `find_packages()`) |
| `install_requires` | Dependencies to install |

## 10. Mini Project: Custom Helper Package

Create a complete, production-ready package with multiple modules:


````python
# Mini Project: Create myhelpers package
print("\n" + "="*70)
print("MINI PROJECT: COMPLETE HELPER PACKAGE")
print("="*70)

# Step 1: Create math_tools module content
print("\nSTEP 1: math_tools.py")
print("-" * 70)

math_tools = '''
"""
math_tools.py - Mathematical utilities
Part of myhelpers package
"""

def add(*numbers):
    """Add multiple numbers"""
    return sum(numbers)

def multiply(*numbers):
    """Multiply multiple numbers"""
    result = 1
    for num in numbers:
        result *= num
    return result

def average(numbers):
    """Calculate average of numbers"""
    if not numbers:
        return 0
    return sum(numbers) / len(numbers)

def factorial(n):
    """Calculate factorial"""
    if n < 0:
        raise ValueError("Factorial not defined for negative numbers")
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

def is_prime(n):
    """Check if number is prime"""
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True
'''

print(math_tools)

# Step 2: Create text_tools module content
print("\nSTEP 2: text_tools.py")
print("-" * 70)

text_tools = '''
"""
text_tools.py - Text manipulation utilities
Part of myhelpers package
"""

def word_count(text):
    """Count words in text"""
    return len(text.split())

def capitalize_words(text):
    """Capitalize first letter of each word"""
    return ' '.join(word.capitalize() for word in text.split())

def reverse_text(text):
    """Reverse text"""
    return text[::-1]

def remove_vowels(text):
    """Remove vowels from text"""
    vowels = "aeiouAEIOU"
    return ''.join(char for char in text if char not in vowels)

def count_letters(text):
    """Count number of letters"""
    return sum(1 for char in text if char.isalpha())

def longest_word(text):
    """Find longest word in text"""
    words = text.split()
    return max(words, key=len) if words else ""
'''

print(text_tools)

# Step 3: Create date_tools module content
print("\nSTEP 3: date_tools.py")
print("-" * 70)

date_tools = '''
"""
date_tools.py - Date and time utilities
Part of myhelpers package
"""

from datetime import datetime, timedelta

def current_date():
    """Get current date"""
    return datetime.now().strftime("%Y-%m-%d")

def current_time():
    """Get current time"""
    return datetime.now().strftime("%H:%M:%S")

def days_between(date1_str, date2_str):
    """Calculate days between two dates"""
    date1 = datetime.strptime(date1_str, "%Y-%m-%d")
    date2 = datetime.strptime(date2_str, "%Y-%m-%d")
    return abs((date2 - date1).days)

def add_days(date_str, days):
    """Add days to date"""
    date = datetime.strptime(date_str, "%Y-%m-%d")
    new_date = date + timedelta(days=days)
    return new_date.strftime("%Y-%m-%d")

def day_of_week(date_str):
    """Get day of week for date"""
    date = datetime.strptime(date_str, "%Y-%m-%d")
    days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
    return days[date.weekday()]
'''

print(date_tools)

# Step 4: Create __init__.py
print("\nSTEP 4: __init__.py")
print("-" * 70)

init_py = '''
"""
myhelpers - A collection of useful helper functions
Version: 1.0.0
"""

__version__ = "1.0.0"
__author__ = "Your Name"

# Import common functions for easy access
from .math_tools import add, multiply, average, factorial, is_prime
from .text_tools import word_count, capitalize_words, reverse_text
from .date_tools import current_date, current_time, days_between

# Define what gets imported with "from myhelpers import *"
__all__ = [
    'add', 'multiply', 'average', 'factorial', 'is_prime',
    'word_count', 'capitalize_words', 'reverse_text',
    'current_date', 'current_time', 'days_between'
]
'''

print(init_py)

# Step 5: Create setup.py
print("\nSTEP 5: setup.py")
print("-" * 70)

setup_py = '''
from setuptools import setup, find_packages

setup(
    name="myhelpers",
    version="1.0.0",
    description="A collection of useful helper functions",
    author="Your Name",
    author_email="your@email.com",
    url="https://github.com/yourname/myhelpers",
    license="MIT",
    packages=find_packages(),
    python_requires=">=3.6",
    classifiers=[
        "Programming Language :: Python :: 3",
        "Programming Language :: Python :: 3.6",
        "Programming Language :: Python :: 3.7",
        "Programming Language :: Python :: 3.8",
        "License :: OSI Approved :: MIT License",
    ],
)
'''

print(setup_py)

# Step 6: Show usage
print("\nSTEP 6: USAGE EXAMPLE - test.py")
print("-" * 70)

usage = '''
# Import from myhelpers package
from myhelpers import add, multiply, average, word_count, current_date

# Use math functions
print(add(2, 3, 4))           # 9
print(multiply(2, 3, 4))      # 24
print(average([10, 20, 30]))  # 20.0

# Use text functions
text = "Hello World"
print(word_count(text))       # 2

# Use date functions
print(current_date())         # 2025-12-06

# Alternative: import modules
from myhelpers import math_tools, text_tools

print(math_tools.factorial(5))  # 120
print(text_tools.reverse_text("Python"))  # nohtyP
'''

print(usage)

print("\nSTEP 7: FOLDER STRUCTURE")
print("-" * 70)

structure = """
myhelpers_project/
â”œâ”€â”€ myhelpers/
â”‚   â”œâ”€â”€ __init__.py
â”‚   â”œâ”€â”€ math_tools.py
â”‚   â”œâ”€â”€ text_tools.py
â”‚   â””â”€â”€ date_tools.py
â”œâ”€â”€ setup.py
â”œâ”€â”€ README.md
â””â”€â”€ test.py
"""

print(structure)

print("\nâœ“ Mini Project Complete!")
print("  Your package is ready to use locally or publish to PyPI!")
````
